Analise de dados

Neste Exemplo vamos analisar este dataset da taxa de ***Zumbificação*** dos clientes do plano de saude da empresa **Umbrella Corporation**



```python
#imporar o pandas
import pandas as pd
```

```python
#Buscando os dados do dataset para o notebook
tabela = pd.read_csv('cancelamentos.csv')
```

```python
display(tabela)
```

Tratamento de dados:

```python
#renomeando tabela cancelou para virou_zumbi
tabela = tabela.rename(columns={'cancelou': 'virou_zumbi'})
display(tabela)
```

```python
#eliminar dados desnecessarios
tabela = tabela.drop('CustomerID', axis=1)
display(tabela)
```

```python
#Verificar dados do dataset
display(tabela.info())
```

```python
#eliminar celulas vazias
tabela = tabela.dropna()
display(tabela.info())
```

```python
#Contagem de zumbis
display(tabela['virou_zumbi'].value_counts())
```

```python
#Verificando a porcentagem de zumbis
display(tabela['virou_zumbi'].value_counts(normalize=True).map('{:2%}'.format))
```

```python
#Verificando a duração dos contratos como segurado
display(tabela['duracao_contrato'].value_counts())
display(tabela['duracao_contrato'].value_counts(normalize=True).map('{:2%}'.format))
```

```python
#correlação de clientes:
correlacao = tabela[[ 'ligacoes_callcenter', 'tempo_como_cliente']].corr()
#quanto mais proximo de 1 maior correlação
display(correlacao.round(2))
```

```python
#remover os contratos mensais
tabela = tabela[tabela['duracao_contrato']!='Monthly']
display(tabela)

#Verificando a duração dos contratos como segurado
display(tabela['duracao_contrato'].value_counts())
display(tabela['duracao_contrato'].value_counts(normalize=True).map('{:2%}'.format))
```
